―	Source three years of price data from a random set of five equities

```python
import pandas as pd
import datetime
import pandas_datareader as pdr
```

```python
stocks = pd.read_csv('constituents_csv.csv')
```

```python
stocks.sample(10)
```

List of S&P 500 Constituents: https://datahub.io/core/s-and-p-500-companies#resource-constituents

```python
# Sampling 5 random stocks from the NASDAQ
import random
stocklist = stocks["Symbol"]
sample = random.choices(stocklist, k=5)
sample
```

```python
start = datetime.datetime(2017,1,2)     
end = datetime.datetime(2019,12,31) 
data = pdr.get_data_yahoo(sample, start, end)

# specifying a more recent 3 year period as new stocks would show as NaN during time periods that are too far back.
```
